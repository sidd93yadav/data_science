```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x,y)
plt.title("X vs Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()





```

```python
import matplotlib.pyplot as plt

# Data
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Explode to highlight the region with the highest revenue
explode = [0.1 if r == max(revenue) else 0 for r in revenue]

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(revenue, labels=regions, autopct='%1.1f%%', startangle=140,
        colors=['skyblue', 'lightgreen', 'orange', 'lightcoral'],
        explode=explode, shadow=True)




# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')

# Show grid for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```
